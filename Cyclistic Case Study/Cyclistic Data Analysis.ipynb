#### Data Manipulation

```python
df1['started_at'] = pd.to_datetime(df1['started_at'])
df1['ended_at'] = pd.to_datetime(df1['ended_at'])
df1
```

```python
df1.dtypes
```

```python
df1['started_date'] = df1['started_at'].dt.date
df1['started_day'] = df1['started_at'].dt.day_name()
df1['started_month'] = pd.DatetimeIndex(df1['started_date']).month_name()
df1.head()
```

```python
starting_time = df1['started_at']#.dt.time
ending_time = df1['ended_at']#.dt.time

df1['ride_duration'] = ending_time - starting_time
df1.head()
```

Let's check if the riders are using the bike more than a day.

```python
df1['Total_duration'] = df1['ended_at'] - df1['started_at']

df1['time_taken'] = df1['Total_duration'] > '23:59:59'

tt_count = df1.groupby(df1['time_taken'], as_index = False)['ride_id'].count()
tt_count
```

And it's obvious that some riders are using the bikes more than a day.

```python
df1.drop(['time_taken'],axis=1, inplace=True)
```

#### Data Visualization

```python
df1['rideable_type'] = df1['rideable_type'].str.rstrip()

# Find value counts
cc = df1['rideable_type'].value_counts()

top = cc.head(10)

plt.figure(figsize=(10, 6))
colors = ['#b20710'] + ['#221f1f'] * (len(top) - 1)
bar_plot = sns.barplot(x=top.index, y=top.values, palette=colors)

plt.xlabel('Member type')
plt.ylabel('Number of Rides')
plt.title('Number of Rides per member-type')

for index, value in enumerate(top.values):
    bar_plot.text(index, value, str(value), ha='center', va='bottom')

plt.show()
```

```python
started_month = df1['started_month'].value_counts()

top_months = started_month.head(5)

#plt.figure(figsize=(16, 8))
colors = ['#b20710'] + ['#221f1f'] * (len(top_months) - 1)
bar_plot = sns.barplot(x=top_months.index, y=top_months.values, palette=colors)

plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Number of rides over months')

# Add count values on top of each bar
for index, value in enumerate(top_months.values):
    bar_plot.text(index, value, str(value), ha='center', va='bottom')

plt.show()
```
